```python
import pandas as pd
```

```python
df=pd.read_csv('data1.csv')
df
```

```python
df.columns
```

```python
df.columns=['state_name','rice_area','rice_prod','wheat_area','wheat_prod','maize_area','maize_prod','chickpea_area','chickpea_prod','sugarcane_area','sugarcane_prod','cotton_area','cotton_prod']
```

```python
df
```

```python
df1=df.groupby(['state_name']).sum()
```

```python
df1.head()
```

```python
df_rice=df1[['rice_area','rice_prod']]
df_wheat=df1[['wheat_area','wheat_prod']]
df_maize=df1[['maize_area','maize_prod']]
df_chickpea=df1[['chickpea_area','chickpea_prod']]
df_sugarcane=df1[['sugarcane_area','sugarcane_prod']]
df_cotton=df1[['cotton_area','cotton_prod']]
```

```python
rice=df_rice.sort_values(by=['rice_prod'],ascending=False).reset_index(drop=False).head().to_dict(orient='records')
wheat=df_wheat.sort_values(by=['wheat_prod'],ascending=False).reset_index(drop=False).head().to_dict(orient='records')
maize=df_maize.sort_values(by=['maize_prod'],ascending=False).reset_index(drop=False).head().to_dict(orient='records')
chickpea=df_chickpea.sort_values(by=['chickpea_prod'],ascending=False).reset_index(drop=False).head().to_dict(orient='records')
sugarcane=df_sugarcane.sort_values(by=['sugarcane_prod'],ascending=False).reset_index(drop=False).head().to_dict(orient='records')
cotton=df_cotton.sort_values(by=['cotton_prod'],ascending=False).head().reset_index(drop=False).to_dict(orient='records')
```

```python
rice
```

```python
import json
def writejson(file_name,chart_data):
    json_ob = json.dumps(chart_data,indent=2)
    with open(file_name,"w") as outfile:
        outfile.write(json_ob)
```

```python
file_list = ['rice.json','wheat.json','maize.json','chickpea.json','sugarcane.json','cotton.json']
dict_list = [rice,wheat,maize,chickpea,sugarcane,cotton]
for i,j in zip(file_list,dict_list):
    writejson(i,j)
```
